# Модули

Любой проект на языке Python, как и сам язык состоит из модулей и
пакетов.  По  мере  роста  программы  появляется  необходимость  в
структурировании кода. Вы можете разделить ее на несколько файлов для
облегчения поддержки и модификаций. Вы также можете легко
переиспользовать функции, которую вы написали в других программах, не
копируя ее определение в каждый файл.

Согласно [документации](https://docs.python.org/3/tutorial/modules.html),
модули это файлы с кодом, куда помещаются определения классов, функций
и переменных, а также операторы  Python.  Имя файла состоит из имени
модуля и расширения `.py`.  Внутри модуля, его имя в виде строки 
храниться в глобальной дандер-переменной с именем `__name__`.

На имена модулей накладываются те же ограничения, что и на любой
идентификатор. Т. е. имя модуля не должно начинаться с цифры и содержать
пробелов, может содержать строчные и заглавные буквы, нижние
подчеркивания и цифры (начиная со второго символа). Например, имя
`foo_bar.py` будет корректным именем модуля, а `foo bar.py` нет. Стоит
упомянуть, что именовать модули по стандарту нужно в стиле `snake_case`.
